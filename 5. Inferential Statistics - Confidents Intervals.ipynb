### Confidence Intervals

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
df = pd.read_csv('coffee_dataset.csv')
df.head()
```

```python
df.shape
```

```python
coffee_full = pd.read_csv('coffee_dataset.csv')
coffee_sample = coffee_full.sample(200)
```

```python
coffee_sample
```

```python
# Explore dataset
coffee_sample.head()
coffee_sample.info()
```

```python
# Proportion of coffee drinkers in sample
mean_drinker = coffee_sample['drinks_coffee'].mean()
mean_non_drinker = 1 - mean_drinker
```

```python
print(mean_drinker)
print(mean_non_drinker)
```

```python
# Average height of coffee drinkers 
drinker_height = coffee_sample[coffee_sample['drinks_coffee']]['height'].mean()
drinker_height
```

```python
# Average height of coffee non drinkers 
non_drinker_height = coffee_sample[~coffee_sample['drinks_coffee']]['height'].mean()
non_drinker_height
```

```python
# Bootstrap sample of 200 draws
bootsamp = coffee_sample.sample(200, replace=True)
bootsamp['drinks_coffee'].mean()
```

```python
# Bootstrap sample of 10,000 times
boot_means = []
for _ in range(10000):
    bootsamp = coffee_sample.sample(200, replace=True) #kalau replace=False chart tidak muncul
    mean = bootsamp[bootsamp['drinks_coffee'] == False]['height'].mean()
    boot_means.append(mean)
    
plt.hist(boot_means, color='salmon')
```

```python
# Obtain 95% confidence interval
np.percentile(boot_means, 2.5), np.percentile(boot_means, 97.5)
```
